```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/lap_times.csv")
```

```python
df.head(10)
```

```python
df = df[["lap","milliseconds"]]
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.shape
```

```python
df = df[df['milliseconds'] < df['milliseconds'].quantile(0.99)]
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['lap'], df['milliseconds'], alpha=0.3)
plt.xlabel("Lap Number")
plt.ylabel("Lap Time(Milliseconds)")
plt.title("Lap Number vs Lap Time")
plt.show()
```
