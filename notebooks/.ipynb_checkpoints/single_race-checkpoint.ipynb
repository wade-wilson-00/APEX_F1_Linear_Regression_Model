# üèéÔ∏è APEX ‚Äì Linear Regression on a Single F1 Race

In this notebook, I apply Linear Regression built **from scratch** using Gradient Descent
to predict Formula 1 lap times for **a single driver in a single race**.

The goal of this notebook is not to achieve perfect predictions,
but to deeply understand:
- how Gradient Descent works,
- why feature scaling matters,
- and why simple models underfit real-world racing data.

Each row represents:

Key columns used:
- `raceId` ‚Üí identifies the race

## üéØ Why Filter to a Single Race and Driver?

Formula 1 data is highly complex:

To make Linear Regression interpretable, I restrict the dataset to:

```python
race_df = df[df['raceId'] == RACE_ID]
driver_df = race_df[race_df['driverId'] == DRIVER_ID]
race_df.head()
driver_df.head()
```

## üß† Feature Selection

Input feature (X):
- `lap` ‚Üí lap number

Target variable (Y):
- `milliseconds` ‚Üí lap time

This creates a **single-feature linear regression** problem:

Lap Time ‚âà w √ó Lap Number + b

This simplicity is intentional to study model behavior and limitations.


```python
X = driver_df['lap'].values.reshape(-1, 1)
Y = driver_df['milliseconds'].values.reshape(-1, 1)
```

## ‚öñÔ∏è Feature Scaling

Gradient Descent is sensitive to feature scale.

Lap numbers are small (1‚Äì70),
while lap times are large (~80,000 ms).

To stabilize learning, I standardize the input feature:

X_scaled = (X ‚àí mean(X)) / std(X)

This ensures:
- smoother convergence
- stable gradients
- comparable parameter updates


```python
X_mean = X.mean()
X_std = X.std()
X_scaled = (X - X_mean) / X_std
```

## üßÆ Cost Function (Mean Squared Error)

The cost function measures how far predictions are from actual values.

Mean Squared Error (MSE) is used:

J(w, b) = (1 / 2m) √ó Œ£ (≈∑ ‚àí y)¬≤

Why squared error?
- Penalizes large mistakes
- Smooth and differentiable
- Works well with Gradient Descent


```python
def cost_function(X_scaled, Y, w, b):
    m = len(Y)
    Y_pred = w * X_scaled + b
    
    cost = (1/(2*m)) * np.sum((Y_pred - Y) ** 2)
    return cost
```

## üîÅ Gradient Descent Optimization

Gradient Descent iteratively updates parameters to minimize cost.

At each iteration:
1. Make predictions using current w and b
2. Compute gradients (direction of steepest increase)
3. Update parameters in the opposite direction

w := w ‚àí Œ± √ó ‚àÇJ/‚àÇw  
b := b ‚àí Œ± √ó ‚àÇJ/‚àÇb

Where Œ± is the learning rate.


```python
def gradient_descent(X_scaled, Y, w, b, learning_rate, iterations):
    m = len(Y)
    cost_history = []

    for i in range(iterations):
        Y_pred = w * X_scaled + b

        dw = (1/m) * np.sum((Y_pred - Y) * X_scaled)
        db = (1/m) * np.sum(Y_pred - Y)

        w = w - learning_rate * dw
        b = b - learning_rate * db

        cost = cost_function(X_scaled, Y, w, b)
        cost_history.append(cost)


        if i % 100 == 0:
            print(f"Iteration {i}, Cost: {cost}")
    return w, b, cost_history
```

## üìâ Model Convergence

During training, the cost is tracked at regular intervals.

A healthy training process shows:
- rapid cost reduction initially
- gradual flattening as convergence is reached

This indicates the model has reached the best solution
possible given the available features.


```python
# Initialize parameters
w = 0.0
b = 0.0

# Train
w, b, cost_history = gradient_descent(
    X_scaled, Y, w, b, learning_rate, iterations
)

# Predict
Y_pred = w * X_scaled + b

print("Final Learned Parameters:")
print(f"w (slope): {w}")
print(f"b (intercept): {b}")

print("\nActual Vs Predicted Lap Time for DriverID - 10")
for i in range(10):
    print(
        f"Lap {int(X[i][0])} ‚Üí "
        f"Actual: {int(Y[i][0])} ms | "
        f"Predicted: {int(Y_pred[i][0])} ms"
    )

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, alpha=0.5, label="Actual")
plt.plot(X, Y_pred, color='red', linewidth=4, label="Predicted")

plt.xlabel("Lap Number")
plt.ylabel("Lap Time (milliseconds)")
plt.title(f"Linear Regression on Single F1 Race {RACE_ID}")
plt.legend()
plt.show()
```

## üîç Making Predictions

After training, predictions are generated using:

≈∑ = w √ó X_scaled + b

These predictions represent the model‚Äôs estimate of lap time
based solely on lap number.

Because only one feature is used, predictions tend to cluster
around the average lap time.


```python
plt.plot(cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Vs Iterations")
plt.show()
```

## üß† Key Observations

- The model underfits the data
- Predictions are smooth and near the mean
- Real lap times fluctuate due to factors not captured by lap number alone

This highlights an important ML lesson:
**feature engineering matters more than algorithm complexity**.
