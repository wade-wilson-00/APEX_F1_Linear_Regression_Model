The objective of this notebook is to **validate my custom Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("../data/raw/lap_times.csv")
df = df[['raceId','driverId','lap','milliseconds']]
df.head()
```

```python
RACE_ID = 354
DRIVER_ID = 4
```

```python
race_df = df[df['raceId'] == RACE_ID]
driver_df = race_df[race_df['driverId'] == DRIVER_ID]
race_df.head()
driver_df.head()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(driver_df['lap'], driver_df['milliseconds'], alpha=0.5)
plt.xlabel("Lap Number")
plt.ylabel("Lap Time (milliseconds)")
plt.title(f"Lap Time Trend for Race {DRIVER_ID}")
plt.show()
```

## ‚öñÔ∏è Critical Insight: Feature Scaling Consistency

Initially, I observed discrepancies between predictions from my model and sklearn.

The issue was **not the algorithm**, but **feature scaling inconsistency**.

Gradient Descent is highly sensitive to feature scale.  
To ensure a fair comparison:

- Both models must be trained on the **same scaled feature space**
- Predictions must also be made using the **same scaled inputs**

Once both models were trained on `X_scaled` instead of raw `X`,
their predictions aligned closely.

This reinforced a key ML principle:

> *Models can only be compared meaningfully when trained and evaluated on identical feature representations.*


```python
X = driver_df['lap'].values.reshape(-1, 1)
Y = driver_df['milliseconds'].values.reshape(-1, 1)

X_mean = X.mean()
X_std = X.std()
X_scaled = (X - X_mean) / X_std
```

```python
def cost_function(X_scaled, Y, w, b):
    m = len(Y)
    Y_pred = w * X_scaled + b
    
    cost = (1/(2*m)) * np.sum((Y_pred - Y) ** 2)
    return cost
```

```python
def gradient_descent(X_scaled, Y, w, b, learning_rate, iterations):
    m = len(Y)
    cost_history = []

    for i in range(iterations):
        Y_pred = w * X_scaled + b

        dw = (1/m) * np.sum((Y_pred - Y) * X_scaled)
        db = (1/m) * np.sum(Y_pred - Y)

        w = w - learning_rate * dw
        b = b - learning_rate * db

        cost_history.append(cost_function(X_scaled, Y, w, b))
    return w, b, cost_history
```

```python
w = 0.0
b = 0.0
learning_rate = 0.01
iterations = 1000
driver_df = driver_df[driver_df['milliseconds'] < 90000]

w, b, cost_history = gradient_descent(
    X_scaled, Y, w, b, learning_rate, iterations
)
```

```python
my_w = w
my_b = b
```

## üìä Prediction Comparison Strategy

After training both models on `X_scaled`, predictions were generated for:

- The same input laps
- The same driver
- The same race

This ensures:
- identical input distribution
- identical prediction targets
- fair, apples-to-apples comparison


```python
model = LinearRegression()
Y_pred_sk = model.fit(X_scaled, Y)
Y_pred_sk = model.predict(X_scaled)
```

```python
print("My Model:")
print(f"w = {my_w}, b = {my_b}")

print("Sklearn:")
print(f"w = {model.coef_[0]}, b = {model.intercept_}")
```

```python
lap_test = np.array([[10]])

my_pred = w * lap_test + b
sk_pred = model.predict(lap_test)

print("My Model prediction:", my_pred[0][0])
print("Sklearn prediction:", sk_pred[0])
```

```python
Y_pred = w * X_scaled + b
```

## üß† Interpretation of Results

Although both models produce nearly identical predictions, they do not fit every
lap perfectly.

This is expected because:
- only a single feature (lap number) is used
- real F1 lap times depend on many hidden variables

The smooth prediction lines represent the **average trend**,
while the scattered actual points reflect real-world race dynamics.

This demonstrates **underfitting**, not model failure.
